```python
#importing libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Probability of Heads
a = np.array([0,1])
choice = []
avg = []
for i in range(1,1000):
    choice.append(np.random.choice(a,1))
    avg.append(np.mean(choice))
plt.plot(range(1,1000),avg)
```
